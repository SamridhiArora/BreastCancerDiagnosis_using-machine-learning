```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Define a helper to extract metrics
def get_metrics(name, y_true, y_pred):
    return {
        "Model": name,
        "Accuracy": accuracy_score(y_true, y_pred),
        "Precision": precision_score(y_true, y_pred),
        "Recall": recall_score(y_true, y_pred),
        "F1 Score": f1_score(y_true, y_pred)
    }

# Collect metrics from each model
results = []
results.append(get_metrics("SVM", y_test, y_pred_svm))
results.append(get_metrics("Random Forest", y_test, y_pred_rf))
results.append(get_metrics("KNN", y_test, y_pred_knn))
results.append(get_metrics("Decision Tree", y_test, y_pred_dt))
results.append(get_metrics("Naive Bayes", y_test, y_pred_nb))
results.append(get_metrics("Logistic Regression", y_test, y_pred_lr))
results.append(get_metrics("AdaBoost", y_test, y_pred_ab))
results.append(get_metrics("Gradient Boosting", y_test, y_pred_gb))
results.append(get_metrics("MLP", y_test, y_pred_mlp))
results.append(get_metrics("Nearest Centroid", y_test, y_pred_ncc))
results.append(get_metrics("Voting Classifier", y_test, y_pred_vc))

# Convert to DataFrame
import pandas as pd
metrics_df = pd.DataFrame(results)

# Display table
metrics_df.sort_values(by="Accuracy", ascending=False, inplace=True)
print(metrics_df)
```

```python
from sklearn.decomposition import PCA
import numpy as np

import numpy as np
cov_matrix = np.cov(X_train_scaled, rowvar=False)
cov_matrix
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
```

💡 What are Eigenvalues and Eigenvectors (in PCA context)?
Eigenvectors = the directions (axes) along which the data spreads most.

👉 These are your new features (principal components).

Eigenvalues = how much spread (variance) is along each eigenvector.
👉 Larger eigenvalue = more important principal component.

```python


# Sort eigenvalues and eigenvectors
sorted_idx = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_idx]
eigenvectors = eigenvectors[:, sorted_idx]

sorted_idx




```

```python
# To retain good variance
explained_variance_ratio = eigenvalues / np.sum(eigenvalues)
cumulative_variance = np.cumsum(explained_variance_ratio)

k_98 = np.argmax(cumulative_variance >= 0.987) + 1
print(f"Components needed for variance: {k_98}")
```

Key Observations & Interpretation
First few PCs capture most of the variance:

PC1 alone explains ~45%

By PC4: ~85% captured

By PC10: ~95% captured

By PC16: ~97.5% (estimated visually)

Curve starts flattening after ~10 components:

Additional components after that contribute very little additional variance

This suggests a lot of redundancy in the original 30 features

At PC30, cumulative variance reaches 1.0:

Because PCA preserves all variance when all components are used (it's just a rotation)

Dimensionality reduction is very effective for this dataset.

We can:

Use 10 components if 95% variance is your target

Use 16 components if you want a bit more variance (~97.5%)

Avoid using all 30, as the last few contribute almost nothing and may only add noise or overfitting risk.

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import pandas as pd

def format_metrics(name, y_true, y_pred):
    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred)
    rec = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    err = 1 - acc
    return {
        "Methods": name,
        "Precision": f"{prec*100:.2f} %",
        "Recall": f"{rec*100:.2f} %",
        "F_Measure": f"{f1*100:.2f} %",
        "Accuracy": f"{acc*100:.2f} %",
        "Error": f"{err*100:.3f} %"
    }

rows = [
    format_metrics("SVM", y_test, y_pred_svm),
    format_metrics("RF", y_test, y_pred_rf),
    format_metrics("KNN", y_test, y_pred_knn),
    format_metrics("DT", y_test, y_pred_dt),
    format_metrics("NB", y_test, y_pred_nb),
    format_metrics("LR", y_test, y_pred_lr),
    format_metrics("AB", y_test, y_pred_ab),
    format_metrics("GB", y_test, y_pred_gb),
    format_metrics("MLP", y_test, y_pred_mlp),
    format_metrics("NCC", y_test, y_pred_ncc),
    format_metrics("VC (LR+SVM)", y_test, y_pred_vote)
]

comparison_df = pd.DataFrame(rows)
print(comparison_df)
```
